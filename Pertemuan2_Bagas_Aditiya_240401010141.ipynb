```python
import pandas as pd
import numpy as np
```

```python
URL = "https://raw.github.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(URL)
df
```

```python
# Q1 : Berapa jumlah baris dan kolom pada dataset Titanic? Kolom apa saja yang tersedia?
print(df.shape, df.columns.tolist())
```

```python
# Q2 : Kolom mana saja yang memiliki missing values? Berapa jumlah dan proporsinya (%)
print(f"Missing value pada dataset: \n {df.isnull().sum()}", 2*'\n', f"Missing value dalam persentase(%): \n {df.isnull().sum()/df.shape[0]*100}")
```

```python
#Q3: Berapa persen penumpang yang selamat (Survived=1)? Gunakan operasi NumPy atau Pandas.
penumpangSelamat = df[df['Survived']==1]
percent = len(penumpangSelamat)/len(df)

print(f"Penumpang selamat :", df['Survived'].sum())
print(f"Persentase selamat (%) : {percent * 100:.2f}%")
```

```python
#Q4: Filter penumpang wanita dari kelas 1. Berapa jumlahnya dan rata-rata usianya?
wk1 = df[(df['Sex'] == 'female') & (df['Pclass'] == 1)]
wk1_survived = wk1[wk1['Survived'] == 1]

print(f"Jumlah wanita di kelas 1 : {wk1.shape[0]} orang")
print(f"Rata-rata usia penumpang wanita yang selamat di Kelas 1 adalah {wk1_survived['Age'].mean():.2f} tahun")
```

```python
#Q5: Gunakan groupby untuk menghitung tingkat keselamatan (mean Survived) per kelas penumpang. Kelas mana yang paling tinggi?
hasil = df.groupby('Pclass').agg(
        total=('PassengerId', 'count'),
        selamat=('Survived','sum'),
        persen = ('Survived','mean'),
).round(3)
print(hasil)
```
